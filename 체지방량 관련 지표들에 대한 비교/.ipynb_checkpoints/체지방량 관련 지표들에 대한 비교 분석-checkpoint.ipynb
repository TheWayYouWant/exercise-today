<!-- Weight(lbs)와 Height(inches)의 단위를 각각 kg, cm으로 변경합니다.

변수들 사이의 상관계수를 조사합니다. -->

```python
data = df.copy()

# 단위를 변경합니다.
data['Weight'] = data['Weight']*0.45
data['Height'] = data['Height']*2.54

# 데이터셋이 너무 크지 않으므로, 모든 특성 간의 표준 상관계수를 corr() 메서드를 이용해 쉽게 계산할 수 있습니다.
data.corr()
```

* 상관관계의 범위는 -1부터 1까지입니다.
* 1에 가까우면 양의 상관관계, -1에 가까우면 강한 음의 상관관계를 가진다는 의미입니다.
* 0에 가까우면 선형적인 상관관계가 없다는 의미입니다.
* 주의! 상관계수는 선형적인 상관관계만 측정합니다. 그렇기에 비선형적인 관계는 잡을 수 없습니다.
* 상관계수는 기울기와 상관이 없습니다. 예를들어, 인치(inch)의 단위의 키는 피트(feet)나 나노미터(nanometer) 단위의 키와 상관계수가 1입니다.

이번에는 상관관계를 시각적으로 확인해보려고 합니다.

* 아래의 그림에서 대각선 방향(왼쪽 위에서 오른쪽 아래로)은 각 변수 자신에 대한 것이라 직선이 되므로, 유용하지 않습니다.
* 그렇기에 이곳에는 각 특성의 히스토그램을 그립니다.

```python
sns.pairplot(data)
```

### 수치와 그림을 통해 다음을 확인할 수 있습니다.

2. Abdomen과 BodyFat또한 높은 양의 상관관계(0.813432)를 가집니다.

# 2. 가설 설정 및 확인하기

비만을 측정하는 정확한 지표에 대한 논란이 존재합니다.

체질량지수(BMI)가 체지방량을 정확히 반영하지 못한다는 보고가 많아, 비만을 측정하는 지표로 적절하지 못하다는 지적이 대표적인 예입니다.

비만을 측정하는 대표적인 지표들은 다음과 같습니다.

### 1. 체질량지수(BMI)

: 체중(kg)을 키(m)의 제곱으로 나누어 구한다.

### 2. 허리둘레(WC)

: 남자는 90cm 이상, 여자는 80cm 이상인 경우 비만으로 분류한다.

### 3. 허리/엉덩이 비(WHR)

: 허리둘레를 엉덩이 둘레로 나누어 구한다.

### 4. 허리/키 비(WHtR)

: 허리둘레를 신장으로 나누어 구한다.
