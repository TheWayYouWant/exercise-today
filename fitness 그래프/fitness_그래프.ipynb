```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
fit = pd.read_csv("fitness analysis.csv")
fit
```

```python
fit.isnull()
```

```python
fit.info()
```

```python
new_cols=['Timestamp','Name','Gender','Age','Exercise_importance','Fitness_level','Regularity','Barriers','Exercises','Do_you','Time','Time_spent','Balanced_diet','prevents_balanced','Health_level','Recommend_fitness','Equipment','Motivation']
```

```python
column_reference=pd.DataFrame(new_cols,fit.columns)
column_reference
```

```python
fit.columns=new_cols
```

```python
#  4개의 기준으로 다 중복되면 제거한다.
fit = fit.drop_duplicates(['Name', 'Gender', 'Age', 'Exercise_importance'])
fit.head()
```

```python
# 그래프로 표현하려고 fit-level 숫자로 바꿔줌
fit.loc[fit['Fitness_level'] == 'Unfit', 'Fitness_level'] = 0
fit.loc[fit['Fitness_level'] == 'Average', 'Fitness_level'] = 1
fit.loc[fit['Fitness_level'] == 'Good', 'Fitness_level'] = 2
fit.loc[fit['Fitness_level'] == 'Very good', 'Fitness_level'] = 3
fit
```

```python
# Regularity에 따른 헬스 레벨
reg = fit.groupby(['Regularity'], as_index=False).mean()
# reg = reg.sort_values(by='Health_level')
reg
```

```python
fig1, ax1 = plt.subplots(figsize=(12, 8))
ax1.plot(reg['Regularity'], reg['Health_level'])
```

```python
# Age에 따른 헬스 레벨
age = fit.groupby(['Age'], as_index=False).mean()
# age = age.sort_values(by='Health_level')
age
```

```python
# 나이가 어리다고 본인이 건강하다고 느끼지 않는다.
fig2, ax2 = plt.subplots(figsize=(12, 8))
ax2.plot(age['Age'], age['Health_level'])
```

```python
# 건강을 얼마나 중요하게 생각하느냐에 따라 본인이 건강하다고 느끼는 지표
exer_important = fit.groupby(['Exercise_importance'], as_index=False).mean()
# exer_important = exer_important.sort_values(by='Health_level')
exer_important
```

```python
fig3, ax3 = plt.subplots(figsize=(12, 8))
ax3.plot(exer_important['Exercise_importance'], exer_important['Health_level'])
```
